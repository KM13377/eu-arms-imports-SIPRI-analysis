```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Country": ["Poland", "Germany", "France", "UK", "Italy"],
    "USA_dependency_2021": [70, 55, 20, 60, 50],
    "USA_dependency_2025": [85, 65, 25, 70, 60]
}

df = pd.DataFrame(data)
df
```

```python
df["change"] = df["USA_dependency_2025"] - df["USA_dependency_2021"]
df.sort_values(by="change", ascending=False)

plt.bar(df["Country"], df["change"])
plt.title("Change in US Arms Dependency (2021-2025)")
plt.ylabel("Percentage point change")
plt.show()
```
